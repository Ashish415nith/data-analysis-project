```python
df=pd.read_csv(url,usecols=["continent",'location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'])
df
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
k=df.dropna(subset=["continent"])
k
```

```python
k.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df["continent"]=df["continent"].fillna(0)
```

```python
df["total_cases"]=df["total_cases"].fillna(0)
```

```python
df["total_deaths"]=df["total_deaths"].fillna(0)
```

```python
df["gdp_per_capita"]=df["gdp_per_capita"].fillna(0)
```

```python
df["human_development_index"]=df["human_development_index"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df.dtypes
```

```python
df["year"]=df["date"].dt.year
df["month"]=df["date"].dt.month
df["day"]=df["date"].dt.day
```

```python
df.head()
```

```python
df.groupby("continent").max().reset_index()
```

```python
df_groupby=df.groupby("continent").max().reset_index()
df_groupby
```

```python
df_groupby["ratio"]=df_groupby["total_deaths"]/df_groupby["total_cases"]
```

```python
df_groupby.head(8)
```

```python

sns.distplot(df_groupby["gdp_per_capita"])
```

```python
sns.jointplot(data=df_groupby,x="total_cases",y="gdp_per_capita",kind="scatter",hue="continent")
```

```python
sns.pairplot(data=df_groupby)
```

```python
plt.figure(figsize=(20,14))
sns.catplot(data=df_groupby,x="continent",y="total_cases",kind="bar")
```

```python
df_groupby.to_csv("df_groupby.csv")
```
